from skimage.io import imread: Baris ini mengimport fungsi imread dari library skimage.io. Fungsi imread digunakan untuk membaca gambar dari file.

from skimage.transform import resize: Baris ini mengimport fungsi resize dari library skimage.transform. Fungsi resize digunakan untuk mengubah ukuran (rescale) gambar.

from skimage.feature import hog: Baris ini mengimport fungsi hog dari library skimage.feature. Fungsi hog (Histogram of Oriented Gradients) digunakan untuk mengekstrak fitur dari gambar dengan tujuan untuk digunakan dalam proses deteksi objek.

from skimage import exposure: Baris ini mengimport sub-module exposure dari library skimage. Sub-module exposure digunakan untuk mengontrol dan meningkatkan kontras gambar.

import matplotlib.pyplot as plt: Baris ini mengimport sub-module pyplot dari library matplotlib dan memberi alias sebagai plt. pyplot digunakan untuk membuat plot dan visualisasi data.

img = imread('kursi.jpg'): Baris ini membaca gambar "kursi.jpg" menggunakan fungsi imread dan menyimpannya dalam variabel img.

plt.axis("off"): Baris ini mengatur agar sumbu pada plot gambar tidak ditampilkan (off), sehingga tidak ada tampilan sumbu x dan y pada gambar saat ditampilkan nanti.

plt.imshow(img): Baris ini menampilkan gambar yang sudah dibaca dan disimpan dalam variabel img menggunakan fungsi imshow dari matplotlib.pyplot.

print(img.shape): Baris ini mencetak bentuk (shape) dari gambar yang sudah dibaca. Shape ini akan menampilkan informasi tentang dimensi gambar, yaitu (tinggi, lebar, jumlah saluran warna).

```python
#IMPORT LIBRARY YANG DIBUTUHKAN
from skimage.io import imread
from skimage.transform import resize
from skimage.feature import hog
from skimage import exposure
import matplotlib.pyplot as plt

# MEMBACA GAMBAR
img = imread('kursi.jpg')
plt.axis("off")
plt.imshow(img)
print(img.shape)
```

resized_img = resize(img, (64*4, 128*4)): Pada baris ini, fungsi resize digunakan untuk merubah ukuran gambar (img) menjadi 4 kali lipat dari ukuran aslinya. Ukuran yang baru adalah (644, 1284), yang berarti lebar gambar akan menjadi 256 piksel (64 * 4) dan tinggi gambar akan menjadi 512 piksel (128 * 4). Proses resizing ini akan mengubah tampilan gambar menjadi lebih besar atau lebih kecil sesuai dengan ukuran yang diinginkan.

plt.axis("off"): Baris ini mengatur untuk tidak menampilkan sumbu koordinat pada plot gambar. Hal ini sering digunakan untuk menyembunyikan sumbu ketika hanya ingin menampilkan gambar tanpa adanya tanda-tanda sumbu.

plt.imshow(resized_img): Pada baris ini, fungsi imshow dari matplotlib.pyplot digunakan untuk menampilkan gambar yang telah diubah ukurannya (resized_img) ke dalam plot.

print(resized_img.shape): Baris ini mencetak bentuk (shape) dari gambar yang telah diubah ukurannya. Dalam konteks ini, resized_img.shape akan mencetak tuple yang berisi (tinggi, lebar, dan jumlah saluran warna). Misalnya, jika gambar berwarna, bentuknya mungkin menjadi (512, 256, 3), yang berarti gambar memiliki tinggi 512 piksel, lebar 256 piksel, dan terdiri dari 3 saluran warna (misalnya merah, hijau, biru).

```python
# MERUBAH UKURAN GAMBAR
resized_img = resize(img, (64*4, 128*4))
plt.axis("off")
plt.imshow(resized_img)
print(resized_img.shape)
```

fd, hog_image = hog(resized_img, orientations=9, pixels_per_cell=(8, 8), cells_per_block=(2, 2), visualize=True, multichannel=True): Pada baris ini, kita menggunakan fungsi hog dari library skimage.feature untuk menghitung Histogram of Oriented Gradients pada gambar resized_img.

resized_img: Gambar yang akan dihitung HOG-nya.
orientations=9: Jumlah bin dalam histogram orientasi. Nilai ini menentukan berapa banyak arah sudut yang akan dihitung di setiap sel (cell) gambar.
pixels_per_cell=(8, 8): Ukuran sel dalam piksel. Ini menentukan berapa banyak piksel yang akan digunakan dalam setiap sel saat menghitung HOG.
cells_per_block=(2, 2): Jumlah sel dalam satu blok. HOG menggabungkan beberapa sel ke dalam blok dan melakukan normalisasi pada setiap blok untuk meningkatkan kestabilan.
visualize=True: Jika bernilai True, fungsi hog akan mengembalikan gambar visualisasi dari HOG yang telah dihitung, yang disimpan dalam variabel hog_image.
multichannel=True: Parameter ini mengindikasikan bahwa gambar yang diberikan memiliki lebih dari satu saluran warna (misalnya, gambar berwarna). Sebagai contoh, gambar RGB memiliki tiga saluran warna (merah, hijau, biru).
plt.axis("off"): Baris ini mengatur untuk tidak menampilkan sumbu koordinat pada plot gambar, mirip seperti penjelasan sebelumnya.

plt.imshow(hog_image, cmap="gray"): Pada baris ini, fungsi imshow dari matplotlib.pyplot digunakan untuk menampilkan gambar visualisasi hasil HOG (hog_image) dalam plot. cmap="gray" menentukan bahwa plot akan ditampilkan dalam skala abu-abu (grayscale) untuk gambar HOG.

```python
fd, hog_image = hog(resized_img, orientations=9, pixels_per_cell=(8, 8),
                    cells_per_block=(2, 2), visualize=True, multichannel=True)
plt.axis("off")
plt.imshow(hog_image, cmap="gray")
```
